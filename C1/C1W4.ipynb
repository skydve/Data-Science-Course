```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
plt.style.use('bmh')
plt.rcParams['figure.figsize'] = (15, 7)
```

## Зададим распредление
Сгенерируем теоретическое распределение хи-квадрат и построим его график

```python
k = 4
rv = sts.chi2(k, 0) #распределение хи-квадрат, МО = k, дисперсия = 2*k
x = np.linspace(0, 30, 1000)
rvs = rv.rvs(1000)
```

```python
plt.hist(rvs, bins=20, normed=True)
plt.plot(x, rv.pdf(x))
plt.ylabel('$f(x)$ and samples hist')
plt.xlabel('$x$')
```

## Построим выборки средних
Для трех значений n сгенерируем 1000 выборок объема n, построим гистограммы их средних. Для каждого графика построим теоретическую функцию нормального распределения по ЦПТ.

```python
for i, n in enumerate([2, 10, 100]):
    a = [np.mean(rv.rvs(n)) for j in range(1000)]
    plt.subplot(1,3,i+1)
    plt.hist(a, normed=True, bins=30, range=(min(a), max(a)))    
    mu = k
    sigma = (2*k/n)**0.5
  
    norm_rv = sts.norm(mu, sigma)
    x = np.linspace(min(a), max(a), 100)
    plt.plot(x, norm_rv.pdf(x))
    plt.ylabel('$f(mean)$')
    plt.xlabel('mu among n = {}'.format(n))
```

## Вывод
Как видно из графиков при увеличении числа элементов выборки (n) выборочное распредление все ближе аппроксимируется к теоретическому нормальному распределению.
